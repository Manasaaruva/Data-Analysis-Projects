```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('E-commerce_sales.csv')
```

```python
dataset.head(2)
```

```python
dataset.shape
```

```python
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'], format='%m/%d/%Y')
```

```python
dataset['order year']=dataset['Order Date'].dt.year
```

```python
# dataset.info()
```

```python
# dataset.isnull().sum()
```

# Order Priority

```python
dataset['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x="Order Priority",data=dataset)
plt.title("Count Of Order Priority")
plt.savefig("Count Of Order Priority.jpg")
plt.show()
```

# Ship Mode

```python
dataset['Ship Mode'].value_counts()
```

```python
x=dataset['Ship Mode'].value_counts().index
y=dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(4,3))
plt.pie(y,labels=x,startangle=90,autopct="%0.2f%%")
plt.legend(loc=3)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode",data=dataset,hue="Category")
plt.show()
```

# Customer Segment

```python
plt.figure(figsize=(4,3))
sns.countplot(x="Segment",data=dataset)
plt.title("Count Of Segment")
plt.show()
```

# Product Category

```python
plt.figure(figsize=(4,3))
sns.countplot(x="Category",data=dataset)
plt.title("Count Of Product Category")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="Category",data=dataset[dataset["Category"]=="Office Supplies"],hue="Sub-Category")
plt.show()
```

# Year Wise

```python
dataset['order year'].value_counts()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x="order year",data=dataset)
plt.title("Count Of Orders Based on Years")
plt.show()
```

# Profit

```python
plt.figure(figsize=(4,3))
sns.barplot(x="Category",y="Profit",data=dataset,estimator='sum')
plt.show()
```
